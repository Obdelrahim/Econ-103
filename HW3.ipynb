```python
import pandas as pd 
import numpy as np 
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("twin_cleaned.csv")
```

```python
df.head()
```

```python
model_ovb = sm.ols(formula = "lwage~educ", data=df)
results_ovb = model_ovb.fit(cov_type = "HC1")
print(results_ovb.summary())
```

```python
model_d = sm.ols(formula = "dlwage~deduc-1", data=df)
results_d = model_d.fit(cov_type = "HC1")
print(results_d.summary())
```

```python
df["educf"] = df["educ"]*df["female"]
```

```python
df.head()
```

```python
model1 = sm.ols(formula = "hrwage~educ + female +educf", data=df)
results1 = model1.fit(cov_type = "HC1")
print(results1.summary())
```

```python
model2 = sm.ols(formula = "hrwage~educ +female + educ:female", data = df)
results2 = model2.fit(cov_type = "HC1")
print(results2.summary())
```

```python
df["pexp"] = df["age"]- df["educ"]- 7
df["educ2"] = (df["educ"])**2
df["pexp2"] = (df["pexp"])**2
df["pexp_educ"] = (df["pexp"])*(df["educ"])
```

```python
main = sm.ols(formula = "hrwage~educ + pexp", data = df)
results_main = main.fit(cov_type = "HC1")
df["resid2"] = (results_main.resid)**2
df.head()
```

```python
aux = sm.ols(formula = "resid2~educ + pexp + educ2 + pexp2 + pexp_educ")
results_aux = aux.fit(cov_type = "HC1")
hypotheses = "(educ=0), (pexp=0), (educ2=0), (pexp2=0), (pexp_educ=0)"
ftest = results_aux.f_test(hypotheses)
```
